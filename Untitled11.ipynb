```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
```

```python

from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("C:/Users/RAJDEEP/Downloads/archive (3)/NIFTY50_all.csv")
```

```python
# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
data = df[['Close']].dropna().sort_index()

# Train-Test Split (80-20)
train_size = int(len(data) * 0.8)
train, test = data.iloc[:train_size], data.iloc[train_size:]

print("Train size:", len(train))
print("Test size:", len(test))
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Close Price')
plt.title("NIFTY 50 Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()

# Rolling mean (trend check)
data['rolling_mean'] = data['Close'].rolling(window=30).mean()

plt.figure(figsize=(12,6))
plt.plot(data['Close'], label="Original")
plt.plot(data['rolling_mean'], label="30-day Rolling Mean")
plt.legend()
plt.show()
```

```python
# Split into train/test
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

print("Train size:", len(train))
print("Test size:", len(test))

```

```python
# Fit ARIMA
arima_model = ARIMA(train['Close'], order=(5,1,0))
arima_fit = arima_model.fit()

# Forecast
forecast_arima = arima_fit.forecast(steps=len(test))

# Evaluation
rmse_arima = np.sqrt(mean_squared_error(test['Close'], forecast_arima))
mape_arima = mean_absolute_percentage_error(test['Close'], forecast_arima)

print("ARIMA RMSE:", rmse_arima)
print("ARIMA MAPE:", mape_arima)

# Plot
plt.figure(figsize=(12,6))
plt.plot(train.index, train['Close'], label="Train")
plt.plot(test.index, test['Close'], label="Test")
plt.plot(test.index, forecast_arima, label="ARIMA Forecast")
plt.legend()
plt.show()
```

```python
# Fit SARIMA
sarima_model = SARIMAX(train['Close'], order=(2,1,2), seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit()

# Forecast
forecast_sarima = sarima_fit.forecast(steps=len(test))

# Evaluation
rmse_sarima = np.sqrt(mean_squared_error(test['Close'], forecast_sarima))
mape_sarima = mean_absolute_percentage_error(test['Close'], forecast_sarima)

print("SARIMA RMSE:", rmse_sarima)
print("SARIMA MAPE:", mape_sarima)

# Plot
plt.figure(figsize=(12,6))
plt.plot(train.index, train['Close'], label="Train")
plt.plot(test.index, test['Close'], label="Test")
plt.plot(test.index, forecast_sarima, label="SARIMA Forecast")
plt.legend()
plt.show()


```

```python
# Use a smaller subset for faster training
train_small = train['Close'].tail(2000)   # last 2000 points
test_small = test['Close'].head(200)      # first 200 test points

# Fit SARIMA with fewer parameters and limited iterations
sarima_model = SARIMAX(train_small, order=(2,1,2), seasonal_order=(0,1,1,12))
sarima_fit = sarima_model.fit(disp=False, maxiter=50)

# Forecast
forecast_sarima = sarima_fit.forecast(steps=len(test_small))

# Evaluation
rmse_sarima = np.sqrt(mean_squared_error(test_small, forecast_sarima))
mape_sarima = mean_absolute_percentage_error(test_small, forecast_sarima)

print("SARIMA RMSE (fast mode):", rmse_sarima)
print("SARIMA MAPE (fast mode):", mape_sarima)

# Plot
plt.figure(figsize=(12,6))
plt.plot(train_small.index, train_small, label="Train (subset)")
plt.plot(test_small.index, test_small, label="Test (subset)")
plt.plot(test_small.index, forecast_sarima, label="SARIMA Forecast")
plt.legend()
plt.show()
```

```python
# Prophet requires 'ds' and 'y' columns
prophet_df = data.reset_index()[['Date','Close']]
prophet_df.columns = ['ds','y']

# Train-Test Split
train_p = prophet_df.iloc[:train_size]
test_p = prophet_df.iloc[train_size:]

# Fit Prophet
model_prophet = Prophet()
model_prophet.fit(train_p)

# Future dataframe
future = model_prophet.make_future_dataframe(periods=len(test_p))
forecast_prophet = model_prophet.predict(future)

# Extract predictions for test
forecast_values = forecast_prophet[['ds','yhat']].tail(len(test_p))

# Evaluation
rmse_prophet = np.sqrt(mean_squared_error(test_p['y'], forecast_values['yhat']))
mape_prophet = mean_absolute_percentage_error(test_p['y'], forecast_values['yhat'])

print("Prophet RMSE:", rmse_prophet)
print("Prophet MAPE:", mape_prophet)

# Plot
model_prophet.plot(forecast_prophet)
plt.show()

# Components
model_prophet.plot_components(forecast_prophet)
plt.show()
```

```python
# Scale data
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data[['Close']])

# Create sequences
def create_sequences(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        X.append(dataset[i:(i+time_step), 0])
        y.append(dataset[i+time_step, 0])
    return np.array(X), np.array(y)

time_step = 60
X, y = create_sequences(scaled, time_step)

# Train-Test split
train_size = int(len(X)*0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape for LSTM
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step,1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# Predict
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1,1))

# Evaluation
rmse_lstm = np.sqrt(mean_squared_error(y_test_actual, y_pred))
mape_lstm = mean_absolute_percentage_error(y_test_actual, y_pred)

print("LSTM RMSE:", rmse_lstm)
print("LSTM MAPE:", mape_lstm)

# Plot
plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label="Actual")
plt.plot(y_pred, label="LSTM Predicted")
plt.legend()
plt.show()
```

```python
# Collect results
models = ['ARIMA','SARIMA','Prophet','LSTM']
rmse_values = [rmse_arima, rmse_sarima, rmse_prophet, rmse_lstm]

# Plot comparison
plt.bar(models, rmse_values, color=['blue','green','orange','red'])
plt.title("Model Comparison (RMSE)")
plt.ylabel("RMSE")
plt.show()

# Print results
for m, r in zip(models, rmse_values):
    print(f"{m} RMSE: {r}")
```
